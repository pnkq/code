```python
import torch
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```python
model_name = "gpt2"
```

```python
sentence = "John loves Mary very much."
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
inputs = tokenizer(sentence, return_tensors="pt")
```

```python
inputs
```

```python
tokens = tokenizer.convert_ids_to_tokens(inputs["input_ids"][0])
```

```python
tokens
```

```python
model = AutoModelForCausalLM.from_pretrained(model_name, output_attentions=True)
```

```python
outputs = model(**inputs)
```

```python
outputs
```

```python
As = outputs.attentions
```

```python
As
```

```python
len(As)
```

```python
A = As[-1] # last attention layer
```

```python
A
```

```python
A.shape # this model has 12 head
```

```python
# get the first head
H = A[0,0].detach().numpy()
```

```python
H
```

```python
# get all 12 heads
Hs = [A[0,h].detach().numpy() for h in range(0,12)]
```

```python
Hs
```

```python
# compute average across all matrices
import numpy as np
```

```python
np.mean(Hs, axis=0)
```
