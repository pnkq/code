```python
import pandas as pd
from bigdl.chronos.data import TSDataset
from sklearn.preprocessing import StandardScaler
```

```python
features = list(df.columns.values)[2:-2] # remove the first two columns and the last two columns which are not the main features ('Date' and 'y')
```

```python
train, valid, test = TSDataset.from_pandas(df, dt_col='Date', target_col='label', extra_feature_col=features, with_split=True, val_ratio=0.1, test_ratio=0.1)
```

```python
valid.df
```

# 1. TCN Forecast

```python
lookback, horizon = 7, 7
```

```python
scaler = StandardScaler()
for data in [train, valid, test]:
    data.deduplicate() \
        .impute() \
        .gen_dt_feature() \
        .scale(scaler, fit=(data is train)) \
        .roll(lookback=lookback, horizon=horizon)
```

```python
X, y = train.to_numpy()
X.shape, y.shape
```

```python
from bigdl.chronos.forecaster.tcn_forecaster import TCNForecaster
```

```python
from torch import nn
```

```python
import torch
```

```python
loss = nn.BCEWithLogitsLoss()
```

```python
model = TCNForecaster(past_seq_len=lookback, future_seq_len=horizon, input_feature_num=X.shape[-1], output_feature_num=y.shape[-1], optimizer='Adam', loss=loss, lr=1e-5)
```

```python
model.fit((X, y), batch_size=32, epochs=10)
```

```python
X_valid, y_valid = valid.to_numpy()
```

```python
model.evaluate((X_valid, y_valid))
```

```python
p_valid_tcn = model.predict(X_valid)
```

```python
import numpy as np
```

```python
def classify(p, epsilon=0.5):
    score = 1 / (1 + np.exp(-p))
    if score >= epsilon:
        return 1.0
    else:
        return 0.0
```

```python
y_v_unscaled = valid.unscale_numpy(y_valid)[:,:,0][:,-1] # after unscaling, values can be double (not 0/1)
```

```python
y_v = [1 if x > 0.9 else 0 for x in y_v_unscaled]
print(y_v)
```

```python
sum(y_v)
```

```python
z_v_tcn = [classify(p) for p in p_valid_tcn[:,:,0][:,-1]]
```

```python
sum(z_v_tcn)
```

```python
import sklearn
import sklearn.metrics
```

```python
# AUC score
fpr, tpr, thresholds = sklearn.metrics.roc_curve(y_v, z_v_tcn)
sklearn.metrics.auc(fpr, tpr)
```

```python
print(sklearn.metrics.classification_report(y_v, z_v_tcn))
```
